### Dot product

```python
# Perpendicular vectors
a = np.array([5, 0])
b = np.array([0, 2])

np.dot(a,b)
```

```python
# Parallel vectors
c = np.array([5, 0])
d = np.array([2, 0])

np.dot(c,d)
```

### Linear transformation

```python
A = np.array([[1, 1], [0, 1]])
B = np.array([[1], [2]])
np.matmul(A, B)
```

```python
np.dot(A,B)
```

```python
A.dot(B)
```

```python
A @ B
```

```python
A = np.array([[1, 2, 3], [3, 4, 5]])
A
```

```python
B = np.array([[2, 2], [4, 4]])
B
```

```python
A.dot(B)
```

```python
A.T.dot(B)
```

### Inverse with linear equation

```python
A = np.array([[1, -1], [1, -2]])
A_inv = np.linalg.inv(A)
B = np.array([[22], [10]])
np.matmul(A_inv, B)
```

```python
A = np.array([[1, -1], [1, -2]])
B = np.array([[22], [10]])
X = np.linalg.solve(A, B)
X
```
